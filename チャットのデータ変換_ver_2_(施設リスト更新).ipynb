## **送迎チャット データ変換↓**

```python
import re
import pandas as pd
from datetime import datetime as dt
import jpholiday

# 非構造化データを読み込む
with open(data_path, 'r', encoding='utf-8') as file:
    data = file.read()

# データを行単位で分割し、空行を削除
lines = [line for line in data.split('\n') if line.strip() != '']

# メッセージの分割
messages = []
current_message = ''
for line in lines:
    if re.match(r'^\s*\d{4}\.\d{2}\.\d{2} \d{2}:\d{2}', line):
        if current_message:
            messages.append(current_message.strip())
        current_message = line.strip()
    else:
        current_message += '\n' + line.strip()
if current_message:
    messages.append(current_message.strip())

# 業務分類のキーワードとカテゴリーの対応（情報共有を最優先）
category_keywords = {
    '情報共有': ['済み', '戻り', '共有', '送迎無し', '送迎関係なし', '送迎関係ありません', '完了', '送迎待機', '洗車', '休憩', '不明', 'IN前', '交通規制', 'もどり', '本日送迎', '送迎不要', '送迎待ちなし', '残り', 'なしで', '無しで', '教えて', '完了', '渋滞', '戻り'],
    '情報共有（鍵）': ['鍵', 'カードキー', 'キー'],
    '本社系': ['売上回収', '売り上げ回収', '施設見学', '月末回収', '見学', '小口配分', '小口配布'],
    '本社系（アメニティ）': ['アメニティ移動'],
    '本社系（スコーン）': ['スコーン'],
    'イレギュラー': ['クレーム', 'ゴキブリ', '汚い', '汚れ', 'エラー', '故障', '虫', '怒り', '怒る', 'イレギュラー', '特別', '緊急', '異常'],
    '設備関係': ['エアコン', 'テレビ', 'トイレ', '洗濯機', 'Wi-Fi', 'リモコン', '充電器', 'キー作成', '鍵作成'],
    '現地チェックイン': ['現地チェックイン', '現地行かれ', '現地直行'],
    'セット数変更': ['セット数変更', '布団追加', '名分追加', '人数変更', '寝具追加', '名追加'],
    '逆ポーター': ['逆ポーター'],
    '自家用車': ['自家', '自身'],
    'ポーター': ['スーツケース', 'カバン', 'バッグ', 'ポータ', 'SC', 'バック', '荷物移動', 'ポーター'],
    'タクシー': ['タクシー', "たくしー"],
    '備品': ['アメニティ', '皿', '鍋', 'ラック', '物干し', '洗剤', '椅子', '机', 'タオルアート', 'コーヒー', 'ベビーベッド', 'アイロンアイロン台', '水', 'グラス', 'アロマ', 'ポーション', 'トイレットペーパー', 'アイロン', 'バスソルト', '歯ブラシ', 'スリッパ', 'バスタオル'],
    '備品（リネン類）': ['タオル', '布団', '掛け布団', 'ベッドパット', 'パット', 'リネン', '寝具', '枕', '敷布団'],
    'イレギュラー（忘れ物）': ['忘れ物', '忘れ'],
    '送迎': ['送迎', 'チェックイン中', 'CI', '名様', 'C/I']
}


# 業務分類を判定する関数（タクシーを情報共有より優先）
def classify_message(content, exclude_info_shared=False):
    # 1. タクシーのキーワードを最初にチェック
    taxi_keywords = category_keywords.get('タクシー', [])
    if any(keyword in content for keyword in taxi_keywords):
        return 'タクシー'

    if not exclude_info_shared:
        # 2. 情報共有のキーワードを次にチェック
        info_shared_keywords = category_keywords.get('情報共有', []) + category_keywords.get('情報共有（鍵）', [])
        if any(keyword in content for keyword in info_shared_keywords):
            return '情報共有'

    # 3. 他のカテゴリーをチェック
    for category, keywords in category_keywords.items():
        if category.startswith('情報共有') or category == 'タクシー':
            continue  # 既に情報共有とタクシーはチェック済み
        if any(keyword in content for keyword in keywords):
            return category
    return '-'

def extract_facility(content):
    # 1. 数字+0+数字のパターンをチェック
    # 例："2009"のような文字列は \d+0\d+ にマッチ
    if re.search(r'\d+0\d+', content):
        return '駅前'

    # 2. テキスト中に '京都駅前' が含まれているかチェック
    kyoto_ekimae_in_content = '京都駅前' in content

    # 3. facility_list を走査して施設名や別名をマッチング
    for facility, aliases in facility_list.items():
        # '京都駅前'が含まれる場合、'駅前' 単体マッチはスキップ
        if facility == '駅前' and kyoto_ekimae_in_content:
            continue

        # 施設名とテキストの一致判定（部分一致含む）
        if facility in content or content in facility:
            return facility

        # 別名（aliases）のチェック
        for alias in aliases:
            # 別名がテキスト内に含まれる、あるいはテキストが別名に含まれる場合
            if alias in content or content in alias:
                return facility

    # どの施設名や別名ともマッチしなかった場合
    return '-'


structured_data = []

for message in messages:
    # 日時、投稿者、内容を抽出
    header_match = re.match(r'^(\d{4}\.\d{2}\.\d{2}) (\d{2}:\d{2})\s+([^\s]+)\s+(.*)', message, re.DOTALL)
    if not header_match:
        continue  # メッセージの形式が正しくない場合はスキップ

    date_str, time_str, sender, content = header_match.groups()
    content = re.sub(r'＠.*', '', content).strip()

    # 確認メッセージを除外
    if 'ご確認' in content and len(content.split('\n')) <= 2:
        continue

    # 投稿者名の修正
    if sender == 'Rinn':
        sender = next((name for name in names_list if name in content), sender)

    # メッセージ時間の取得
    message_hour = int(time_str.split(':')[0])

    # 「戻り次第」の処理
    exclude_info_shared = False
    if '戻り次第' in content:
        content = content.replace('戻り次第', '').strip()
        exclude_info_shared = True

    # 「予定」の処理
    if '予定' in content:
        lines = content.split('\n')
        schedule_index = next((idx for idx, line in enumerate(lines) if '予定' in line), -1)
        if schedule_index != -1:
            for task_line in lines[schedule_index+1:]:
                if not task_line.strip() or re.match(r'^\s*\d{4}\.\d{2}\.\d{2} \d{2}:\d{2}', task_line):
                    break

                # タスク行を処理
                task_content = task_line.strip()

                # 「戻り次第」の処理をタスクにも適用
                if '戻り次第' in task_content:
                    task_content = task_content.replace('戻り次第', '').strip()
                    task_exclude_info_shared = True
                else:
                    task_exclude_info_shared = False

                # 施設名の抽出
                facility = extract_facility(task_content)

                # 時間表記パターン (HH:MM または 数字+時)
                # 例：08:30, 9:00, 10:00, 8時, 10時, 9時など
                time_pattern = r'\b\d{1,2}[:：]\d{2}\b|\b\d+時(?:半)?\b'

                # タスク内容に施設名があり、かつ時間が(HH:MM形式 or 数字+時)が含まれている場合は業務分類を情報共有とする
                if facility != '-' and re.search(time_pattern, task_content):
                    category = '情報共有'
                else:
                    # 通常の業務分類の判定
                    category = classify_message(task_content, exclude_info_shared=task_exclude_info_shared)

                # 時間が16時以降で、内容に特定のキーワードが含まれる場合、業務分類を「送迎」に設定（「情報共有」ではない）
                if (message_hour >= 16 and
                    any(keyword in task_content for keyword in ['スーツケース', 'バッグ', 'SC']) and
                    not any(keyword in task_content for keyword in ['ポーター', 'ポータ']) and
                    category != '情報共有'):
                    category = '送迎'
                elif (category == '-' and
                      message_hour >= 16 and
                      len(task_content) <= 20 and
                      facility != '-' and
                      category != '情報共有'):
                    category = '送迎'

                # ポーターの特別処理（必要なら従来通り）
                if category == 'ポーター':
                    # 内容から施設名とその位置を抽出
                    facilities_in_content = []
                    for facility_name in facility_list:
                        if facility_name == '駅前' and '京都駅前' in task_content:
                            continue
                        for match in re.finditer(re.escape(facility_name), task_content):
                            facilities_in_content.append((match.start(), facility_name))
                    # 数字+0+数字の構造をチェックし、'駅前'を追加
                    if re.search(r'\d+0\d+', task_content):
                        facilities_in_content.append((len(task_content), '駅前'))

                    # 位置でソート
                    facilities_in_content.sort()
                    # 重複を削除
                    facilities_in_content = list(dict.fromkeys(facilities_in_content))
                    if len(facilities_in_content) >= 2:
                        for i in range(len(facilities_in_content)):
                            current_facility = facilities_in_content[i][1]
                            if i < len(facilities_in_content) - 1:
                                next_facility = facilities_in_content[i+1][1]
                                current_pos = facilities_in_content[i][0]
                                next_pos = facilities_in_content[i+1][0]
                                content_between = task_content[current_pos + len(current_facility):next_pos].strip()
                            else:
                                content_between = task_content[facilities_in_content[i][0] + len(facilities_in_content[i][1]):].strip()

                            if '戻り次第' in content_between:
                                content_between = content_between.replace('戻り次第', '').strip()
                                content_between_exclude_info_shared = True
                            else:
                                content_between_exclude_info_shared = False

                            category = classify_message(content_between, exclude_info_shared=content_between_exclude_info_shared)

                            if (message_hour >= 16 and
                                any(keyword in content_between for keyword in ['スーツケース', 'バッグ', 'SC']) and
                                not any(keyword in content_between for keyword in ['ポーター', 'ポータ']) and
                                category != '情報共有'):
                                category = '送迎'
                            elif (category == '-' and
                                  message_hour >= 16 and
                                  len(content_between) <= 20 and
                                  current_facility != '-' and
                                  category != '情報共有'):
                                category = '送迎'

                            structured_data.append({
                                '日付': date_str,
                                '時間': time_str,
                                '投稿者': sender,
                                '施設': current_facility,
                                '業務分類': category,
                                '内容': content_between
                            })
                        continue

                # 通常のデータ追加
                structured_data.append({
                    '日付': date_str,
                    '時間': time_str,
                    '投稿者': sender,
                    '施設': facility,
                    '業務分類': category,
                    '内容': task_content
                })
        continue

    # 通常のメッセージ処理
    facility = extract_facility(content)

    # 業務分類の判定
    category = classify_message(content, exclude_info_shared=exclude_info_shared)

    # 時間が16時以降で、内容に特定のキーワードが含まれる場合、業務分類を「送迎」に設定（「情報共有」ではない）
    if (message_hour >= 16 and
        any(keyword in content for keyword in ['スーツケース', 'バッグ', 'SC']) and
        not any(keyword in content for keyword in ['ポーター', 'ポータ']) and
        category != '情報共有'):
        category = '送迎'
    elif (category == '-' and
          message_hour >= 16 and
          len(content) <= 20 and
          facility != '-' and
          category != '情報共有'):
        category = '送迎'

    # ポーターの特別処理
    if category == 'ポーター':
        # 内容から施設名とその位置を抽出
        facilities_in_content = []
        for facility_name in facility_list:
            if facility_name == '駅前' and '京都駅前' in content:
                continue
            for match in re.finditer(re.escape(facility_name), content):
                facilities_in_content.append((match.start(), facility_name))
        # 数字+0+数字の構造をチェックし、'駅前'を追加
        if re.search(r'\d+0\d+', content):
            facilities_in_content.append((len(content), '駅前'))

        # 位置でソート
        facilities_in_content.sort()
        # 重複を削除
        facilities_in_content = list(dict.fromkeys(facilities_in_content))
        if len(facilities_in_content) >= 2:
            for i in range(len(facilities_in_content)):
                current_facility = facilities_in_content[i][1]
                if i < len(facilities_in_content) - 1:
                    next_facility = facilities_in_content[i+1][1]
                    current_pos = facilities_in_content[i][0]
                    next_pos = facilities_in_content[i+1][0]
                    content_between = content[current_pos + len(current_facility):next_pos].strip()
                else:
                    content_between = content[facilities_in_content[i][0] + len(facilities_in_content[i][1]):].strip()

                if '戻り次第' in content_between:
                    content_between = content_between.replace('戻り次第', '').strip()
                    content_between_exclude_info_shared = True
                else:
                    content_between_exclude_info_shared = False

                category = classify_message(content_between, exclude_info_shared=content_between_exclude_info_shared)

                if (message_hour >= 16 and
                    any(keyword in content_between for keyword in ['スーツケース', 'バッグ', 'SC']) and
                    not any(keyword in content_between for keyword in ['ポーター', 'ポータ']) and
                    category != '情報共有'):
                    category = '送迎'
                elif (category == '-' and
                      message_hour >= 16 and
                      len(content_between) <= 20 and
                      current_facility != '-' and
                      category != '情報共有'):
                    category = '送迎'

                structured_data.append({
                    '日付': date_str,
                    '時間': time_str,
                    '投稿者': sender,
                    '施設': current_facility,
                    '業務分類': category,
                    '内容': content_between
                })
            continue

    # 通常のデータ追加
    structured_data.append({
        '日付': date_str,
        '時間': time_str,
        '投稿者': sender,
        '施設': facility,
        '業務分類': category,
        '内容': content
    })

# データをDataFrameに変換
df = pd.DataFrame(structured_data)

# 「データソース」列を追加し、「内容」列の前に挿入
df.insert(df.columns.get_loc('内容'), 'データソース', '送迎チャット')

# 曜日計算と祝日判定
weekday_mapping = {0: '月', 1: '火', 2: '水', 3: '木', 4: '金', 5: '土', 6: '日'}

df['曜日'] = pd.to_datetime(df['日付']).apply(lambda date: weekday_mapping[date.weekday()])
df['祝日'] = pd.to_datetime(df['日付']).apply(lambda date: '祝日' if jpholiday.is_holiday(date) else '')

# 「時間」列の後ろに「曜日」を挿入
weekday_col = df.pop('曜日')
df.insert(df.columns.get_loc('時間') + 1, '曜日', weekday_col)

# 曜日列の後ろに「祝日」列を挿入
holiday_col = df.pop('祝日')
df.insert(df.columns.get_loc('曜日') + 1, '祝日', holiday_col)

# Excelファイルとして保存
df.to_excel(save_path, index=False)
print(f"構造化データをExcelファイルとして保存しました: {save_path}")

# 最初の15行を表示
print(df.head(15))
```

```python
import re
import pandas as pd
import jpholiday

def extract_facility(content):
    # 1. 数字+0+数字のパターンをチェック
    # 例："2009"のような文字列は \d+0\d+ にマッチ
    if re.search(r'\d+0\d+', content):
        return '駅前'

    # 2. テキスト中に '京都駅前' が含まれているかチェック
    kyoto_ekimae_in_content = '京都駅前' in content

    # 3. facility_list を走査して施設名や別名をマッチング
    for facility, aliases in facility_list.items():
        # '京都駅前'が含まれる場合、'駅前' 単体マッチはスキップ
        if facility == '駅前' and kyoto_ekimae_in_content:
            continue

        # 施設名が含まれるか、またはその逆（部分一致含む）
        if facility in content or content in facility:
            return facility

        # 別名（aliases）のチェック
        for alias in aliases:
            if alias in content or content in alias:
                return facility

    # どの施設名や別名ともマッチしなかった場合
    return '-'

# 非構造化データを読み込む
with open(data_path, 'r', encoding='utf-8') as file:
    data = file.read()

# メッセージを個別に分割
messages = re.findall(r'(\d{4}\.\d{2}\.\d{2} \d{2}:\d{2}  .+?)(?=\n \d{4}\.\d{2}\.\d{2} \d{2}:\d{2}|$)', data, re.DOTALL)

structured_data = []

for message in messages:
    # 日時、投稿者、内容を抽出
    header_match = re.match(r'(\d{4}\.\d{2}\.\d{2} \d{2}:\d{2})\s+([^\n]+?)\s+(.+)', message, re.DOTALL)
    if header_match:
        datetime_str = header_match.group(1).strip()
        sender = header_match.group(2).strip()
        content = header_match.group(3).strip()

        # 日付と時間を分割
        try:
            date_str, time_str = datetime_str.split(' ', 1)
        except ValueError:
            # 日時のフォーマットが正しくない場合はスキップ
            continue

        # 投稿者名の修正
        if sender == 'Rinn':
            sender = next((name for name in names_list if name in content), sender)

        # 空行を削除
        content = '\n'.join([line.strip() for line in content.splitlines() if line.strip()])

        # `＠` 後の内容を削除
        content = re.sub(r'＠.*', '', content)

        # 「お願い」と「お疲れ様です」を含む行を削除
        content = '\n'.join(
            [line for line in content.splitlines() if 'お疲れ様です' not in line]
        )

        # 施設名を抽出 (新たな関数を使用)
        facility = extract_facility(content)

        # 内容が単なる確認のメッセージの場合は除外
        if 'ご確認' in content and len(content.split('\n')) <= 2:
            content_text = ''
        else:
            content_text = content

        # 業務分類を判定
        if any(keyword in content for keyword in ['エアコン', 'テレビ', 'トイレ', '洗濯機', 'Wi-Fi', '故障', 'リモコン', '充電器']):
            category = '設備関係'
        elif any(keyword in content for keyword in ['共有', '完了']):
            category = '情報共有'
        elif any(keyword in content for keyword in ['タオル', '布団', '掛け布団', 'ベッドパット', 'パット', 'リネン']):
            category = '備品（リネン類）'
        elif any(keyword in content for keyword in ['アメニティ', "傘", 'コーヒー', '洗剤', '水', 'グラス', 'アロマ']):
            category = '備品'
        elif '逆ポーター' in content:
            category = '荷物（逆ポーター）'
        elif any(keyword in content for keyword in ['スーツケース', 'バッグ']):
            category = '荷物'
        elif any(keyword in content for keyword in ['セット数変更', '布団追加', '人数変更', '寝具追加']):
            category = 'セット数変更'
        elif any(keyword in content for keyword in ['忘れ物', '忘れ']):
            category = 'イレギュラー（忘れ物）'
        elif any(keyword in content for keyword in ['クレーム', 'エラー', '故障', '虫', '怒り', '怒る']):
            category = 'イレギュラー'
        elif any(keyword in content for keyword in ['鍵', "カードキー"]):
            category = '情報共有（鍵）'

        else:
            category = '-'

        # 単なる確認のメッセージの場合、内容を空にする
        if content_text == '':
            continue

        # 構造化データに追加
        structured_data.append({
            '日付': date_str,
            '時間': time_str,
            '投稿者': sender,
            '施設': facility,
            '業務分類': category,
            '内容': content_text
        })

# データフレームに変換
df = pd.DataFrame(structured_data)

# 「データソース」列を追加し、「内容」列の前に挿入
df.insert(df.columns.get_loc('内容'), 'データソース', '町家管理チャット')

# 曜日計算と祝日判定
weekday_mapping = {0: '月', 1: '火', 2: '水', 3: '木', 4: '金', 5: '土', 6: '日'}

df['曜日'] = pd.to_datetime(df['日付']).apply(lambda date: weekday_mapping[date.weekday()])
df['祝日'] = pd.to_datetime(df['日付']).apply(lambda date: '祝日' if jpholiday.is_holiday(date) else '')

# 「時間」列の後ろに「曜日」を挿入
weekday_col = df.pop('曜日')
df.insert(df.columns.get_loc('時間') + 1, '曜日', weekday_col)

# 曜日列の後ろに「祝日」列を挿入
holiday_col = df.pop('祝日')
df.insert(df.columns.get_loc('曜日') + 1, '祝日', holiday_col)

# Excelファイルとして保存
df.to_excel(save_path, index=False)
print(f"構造化データをExcelファイルとして保存しました: {save_path}")

# 最初の15行を表示
print(df.head(15))
```
